```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("train.csv")
```

```python
data.groupby('State')['Sales'].sum().plot(kind='bar')
plt.title("Sales by State")
plt.show()
```

```python
data['Sales'].plot(kind='hist')
plt.title("Sales Distribution")
plt.show()
```

```python
data.groupby('Country')['Sales'].sum().plot(kind='bar')
plt.title("Sales by City")
plt.show()
```

```python
plt.scatter(data['Sales'], data['State'])
plt.title("Sales vs State")
plt.xlabel("Sales")
plt.ylabel("State")
plt.show()
```

```python
top_states = data.groupby('State')['Sales'].sum().nlargest(5).index

filtered_data = data[data['State'].isin(top_states)]

plt.scatter(filtered_data['Sales'], filtered_data['State'])
plt.title("Top 5 States by Sales")
plt.xlabel("Sales")
plt.ylabel("State")
plt.show()
```

```python
import seaborn as sns

sns.boxplot(x=data['Sales'])
plt.show()
```

```python
sns.histplot(data['Sales'])
plt.show()
```

```python
sns.pairplot(data)
plt.show()
```

```python
import plotly.express as px
data = pd.read_csv("train.csv")
```

```python
fig = px.bar(data, x='Category', y='Sales')
fig.show()
```

```python
fig = px.bar(data, x='Category', y='Sales', color='Category')
fig.show()
```

```python
fig = px.pie(data, names='Category', values='Sales')
fig.show()
```

```python
fig = px.line(data, x='Order Date', y='Sales')
fig.show()
```

```python
fig = px.scatter(data, x='Sales', y='Row ID', color='Category')
fig.show()
```

```python
fig = px.bar(data, x='Category', y='Sales', color='Region')
fig.show()
```

```python
grouped = data.groupby('Category')['Sales'].sum().reset_index()

fig = px.bar(grouped, x='Category', y='Sales')
fig.show()
```
