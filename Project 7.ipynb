```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import load_dataset
#titanic dataset
data = pd.read_csv("Titanic-Dataset.csv")
#tips dataset
tips = load_dataset("tips")
```

```python
print(data)
```

```python
sns.countplot(data['Survived'])
plt.show()
```

```python
data['Sex'].value_counts().plot(kind="pie", autopct="%.2f")
plt.show()
```

```python
plt.hist(data['Age'], bins=5)
plt.show()
```

```python
sns.distplot(data['Age'])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x=tips["total_bill"], y=tips["tip"])

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(
    x=tips["total_bill"],
    y=tips["tip"],
    hue=tips["sex"]
)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(
    x=tips["total_bill"],
    y=tips["tip"],
    hue=tips["sex"],
    style=tips["smoker"]
)

plt.show()
```
