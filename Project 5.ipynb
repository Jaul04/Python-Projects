```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("customer_data.csv")
```

```python
print(data)
```

```python
print(data.columns)
```

```python
print(data.info())
```

```python
print(data.isnull().sum())
```

```python
data = data.dropna()
```

```python
print(data)
```

```python
import numpy as np

print("Average Sales:", np.mean(data['age']))
print("Max Profit:", np.max(data['customer_id']))
print("Min Profit:", np.min(data['gender']))
```

```python
cat_profit = data.groupby('customer_id')['age'].sum()
print(cat_profit)
```

```python
import matplotlib.pyplot as plt

cat_profit.plot(kind='bar')
plt.title("Cust vs Age")
plt.show()
```
