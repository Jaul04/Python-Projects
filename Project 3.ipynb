```python
data = {
    'NAME' : ['Jaul','Amir','Kuldeep','Sidhhu','Shekhar','Ketan','Omkar','Arya'] ,
    'MARKS' : [80, 75, 90, 70, 40, 50, 0, 20] , 
    'GENDER' : ['M', 'M', 'M', 'M', 'M', 'M', 'M', 'M']
}
```

```python
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
print(df.isnull())
```

```python
print(df.describe())
```

```python
df['MARKS'] = df['MARKS'].fillna(df['MARKS'].mean())
print(df)
```

```python
df['GENDER'] = df['GENDER'].map({'M': 0, 'F': 1})
print(df)
```

```python
df = df[df['MARKS'] > 75]
print(df)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'NAME' : ['Jaul','Amir','Kuldeep','Sidhhu','Shekhar','Ketan','Omkar','Arya'],
    'MARKS' : [80, 75, 90, 70, 40, 50, 0, 20],
    'GENDER' : ['M', 'M', 'M', 'M', 'M', 'M', 'M', 'M']
}

df = pd.DataFrame(data)
```

```python
sns.pairplot(df)
plt.show()
```

```python
df_sorted = df.sort_values(by='MARKS', ascending=False)

sns.barplot(x='NAME', y='MARKS', data=df_sorted)
plt.xticks(rotation=45)
plt.show()
```
